### Описание производственного процесса

```python
Image(url="https://metallplace.ru/upload/metphoto/kovshpech.jpg")
```

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

## Создание признаков для обучения

Для создания единой таблицы нам нужно объединить все нужные данные. Каждая строчка в этой таблице будет представлять собой наблюдение об одной партии. Множественные наблюдения об одной партии по одному признаку мы будем агрегировать.

### Дополнительное очистка данных о температуре

В таблице df_temp у нас находятся как начальные температуры(те с которыми сталь только заливалась в ковш), так и температуры после всех добавок и нагреваний - то есть наш целевой признак.

Начальная температура по идее самый важный признак для обучения модели, поэтому отказоваться от него было бы неразумно. Поэтому сперва очистим данные о температуре от тех строк, в которых представленны только единичные замеры.

```python
#найдем уникальные значения столбца key
uniq_temp_keys = df_temp['key'].value_counts()

#удалим из df_temp те партии, для которых есть записи только об одном измерении
df_temp_clear = df_temp[~df_temp['key'].isin(uniq_temp_keys[uniq_temp_keys == 1].index)]
#проверим
df_temp_clear['key'].value_counts().tail()
```

### Разбиение данных о температуре

Сначала создадим таблицу с начальной температурой, которую мы будем использовать как обучающий признак.

```python
#группируем партии и выбираем самое ранее время
initial_temp = pd.DataFrame(df_temp_clear.groupby('key')['Время замера'].min())

#по выбрамонному времени присоединяем значения температур
initial_temp = initial_temp.merge(df_temp_clear, on='Время замера', how='left') 
initial_temp.columns = ['Время первого замера', 'key', 'Начальная t']
initial_temp.head()
```

Теперь создадим таблицу с последними замерами температур - целевой признак.

```python
#группируем партии и выбираем самое ранее время
last_temp = pd.DataFrame(df_temp_clear.groupby('key')['Время замера'].max())

#по выбрамонному времени присоединяем значения температур
last_temp = last_temp.merge(df_temp_clear, on='Время замера', how='left') 
last_temp.columns = ['Время последнего замера', 'key', 'Конечная t']
last_temp.head()
```

Соеденим полученные таблицы.

```python
temp_final = initial_temp.merge(last_temp, on='key', how='inner')
temp_final.head()
```

### Добавление времени обработки стали

Не имеет смысла исползовать время начала и конца обработки стали как обучающий признак, однако промежуток времени был бы неплохим признаком. Добавим его.

```python
temp_final['Время обработки'] = (temp_final['Время последнего замера'] - temp_final['Время первого замера']
                                ).dt.total_seconds()
temp_final.head()
```

Посмотрим на распределение нового признака.

```python
plt.figure(figsize=(8, 5))
temp_final['Время обработки'].hist(bins=100)
plt.title('Распределение количества секунд на нагрев')
plt.ylabel('количество наблюдений')
plt.xlabel('секунды')
plt.show()
```

```python
temp_final['Время обработки'].describe()
```

На гистограмме явно видно, что есть аномальные значения. Это так же подтвержает сильное отличие медианы от максимального. После 5000 секунд значений почти нет, посмотрим сколько их точно.

```python
temp_final[temp_final['Время обработки'] > 5000].shape[0]
```

Их мало. Безусловно такое возможно в технологическом процессе, но с целью усреднения данных и повышения качества модели от выбросов лучше избавиться.

```python
temp_final = temp_final[temp_final['Время обработки'] < 5000]
```

### Объединение таблиц в одну

За основу возьмем уже созданную таблицу *temp_final*. К ней будем добавлять остальные данные, связывая их по номеру партии.

Первой будем присоединять информацию об электродах.

Для обучения из двух мощностей будем брать только *активную*. По сути только активная мощность идет на нагревание, а реактивная уходит обратно в сеть. Да и к тому же между ними очень сильная корреляция, удаление одной поможет избежать мультиколлинеарности. 

Так как каждая партия неоднократно нагревалась, то в таблицу пойдет суммарная активная мощность по партиям.

```python
# агрегируем данные по активной мощности
df_arc_agg = df_arc[['key', 'Активная мощность']].groupby('key').sum()
```

Теперь соединим все вместе.

Заказчик, в лице металлургического комбината *ООО «Так закаляем сталь»*, решил уменьшить потребление электроэнергии на этапе обработки стали.
